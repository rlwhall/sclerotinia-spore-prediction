```python
cats = df.select_dtypes('object').columns.tolist()
```

```python
cont = df.drop(labels=cats, axis=1)
```

```python
cont.columns
```

```python
def create_cat(data):
    if data < 200:
        return 0
    else:
        return 1
```

```python
cont['label'] = cont['SsMean'].apply(create_cat)
```

```python
print(cont.loc[cont['label'] == 0].label.count())
print(cont.loc[cont['label'] == 1].label.count())
```

```python
X = cont.drop(labels=['SsMean', 'logSsMean','logSsMean_t1', 'label'], axis=1)
y = cont.label
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=233)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.feature_selection import SelectKBest

kbest = SelectKBest(k=20)
kbest = kbest.fit(X_train,y_train)
X2_train = kbest.transform(X_train)
X2_test = kbest.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
model = lr.fit(X2_train,y_train)
```

```python
y_pred = lr.predict(X2_test)
lr.score(X2_test, y_test)
```

```python
from sklearn.metrics import roc_auc_score

probs = lr.predict_proba(X2_test)[:,1]
auc = roc_auc_score(y_test,probs)
auc
```

```python
from sklearn.metrics import classification_report

cr = classification_report(y_test,y_pred)
print(cr)
```
