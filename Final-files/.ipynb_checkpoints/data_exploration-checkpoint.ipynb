## Data exploration and Visualization

```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load processed spore and provincial weather data
df = pd.read_csv('SFE_data.csv', index_col=0)
```

```python
# describe target variable
df.SsMean.describe()
```

```python
df.shape
```

```python
# spore counts by location
locations = df.FieldID.unique().tolist()

#note: not all years have same day range
for loc in locations:
    # select July JDay's so that x axis is same for all plots
    pts = df[df['FieldID'] == loc]
    plt.plot(pts.SsMean)
    plt.xlabel('Day of Year')
    plt.ylabel('Daily Spore Count')
    plt.title('Field ID: {}'.format(loc))
    plt.ylim(-300,25000)
    plt.show()
```

```python
# average daily spore counts by year week
years = df.year.unique().tolist()

for year in years:
    # filter to makes sure x-axis is same
    pts = df[(df['year'] == year) & (df['YearWeek'] >= 27) & (df['YearWeek'] <=35)].groupby(by='YearWeek').SsMean.mean()
    plt.plot(pts.index, pts.values)
    plt.title('Average Daily Spore Count by Week')
    plt.legend(labels = years)
    plt.ylabel('Average Spore Count')
    plt.xlabel('Year Week')
```

```python
# spore counts by month
pts = df[['month', 'SsMean']]
sns.set(rc={'figure.figsize': (7, 7)})
sns.boxplot(data=pts, x='month', y='SsMean', showfliers=False).set_title('Mean Spore Count by Month')
```

```python
# spore counts by year
pts = df[['year', 'SsMean']]
sns.set(rc={'figure.figsize': (7, 7)})
sns.boxplot(data=pts, x='year', y='SsMean', showfliers=False).set_title('Mean Spore Count by Year')
```

```python
# look at correlations
to_corr = ['SsMean', 'Crop', 'MinTemp_y', 'MaxTemp_y',
       'MeanTemp_y', 'MeanRH_y', 'Precip', 'Wind2', 'Wind10', 'Wind10Dir',
       'Frost', 'MeanDP_y', 'MinTemp_uDP', 
       'MaxTemp_u20',  'MeanTemp_u20', 'MinTemp_1', 'MinTemp_3',
       'MinTemp_7', 'MaxTemp_1', 'MaxTemp_3', 'MaxTemp_7', 'MeanTemp_1',
       'MeanTemp_3', 'MeanTemp_7', 'MeanRH_1', 'MeanRH_3', 'MeanRH_7',
       'Precip_1', 'Precip_3', 'Precip_7', 'Wind2_1', 'Wind2_3', 'Wind2_7',
       'Wind10_1', 'Wind10_3', 'Wind10_7', 'Wind10Dir_1', 'Wind10Dir_3',
       'Wind10Dir_7', 'Frost_1', 'Frost_3', 'Frost_7', 'MeanDP_1', 'MeanDP_3',
       'MeanDP_7', 'PrecipSum_1', 'PrecipSum_3', 'PrecipSum_7',
       'MinTemp_uDPSum_1', 'MinTemp_uDPSum_3', 'MinTemp_uDPSum_7',
       'MaxTemp_u20Sum_1', 'MaxTemp_u20Sum_3', 'MaxTemp_u20Sum_7',
       'MeanTemp_u20Sum_1', 'MeanTemp_u20Sum_3', 'MeanTemp_u20Sum_7', 'month',
       'YearWeek']

corr = df[to_corr].corr()
print(corr['SsMean'].sort_values().head())
print(corr['SsMean'].sort_values().tail())
```

```python
# select 10 most correlated features
ab_vals = abs(corr['SsMean'].copy())
pl1 = ab_vals.sort_values(ascending=False).head(11)

# plot
fig, ax = plt.subplots()
fig = sns.barplot(x=pl1.index[1:], y=pl1.values[1:], color='b')
plt.ylim(0,1)
plt.xticks(rotation=80)
ax.set_xticklabels(labels=pl1.index.tolist()[1:])
plt.show()
```
